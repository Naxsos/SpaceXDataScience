# **Launch Sites Locations Analysis with Folium**


The launch success rate may depend on many factors such as payload mass, orbit type, and so on. It may also depend on the location and proximities of a launch site, i.e., the initial position of rocket trajectories. Finding an optimal location for building a launch site certainly involves many factors and hopefully we could discover some of the factors by analyzing the existing launch site locations.


### Content of the Notebook

*   All launch sites marked on a map
*   Success/failed launches for each site marked on the map
*   Calculated distances between a launch site to its proximities

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# find coordinate of the closet coastline
launch_site_lat = 28.56327
launch_site_lon = -80.57673
coastline_lat = 28.56315
coastline_lon = -80.56798

distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
coordinate = [coastline_lat,coastline_lon] 
distance_marker = folium.Marker(
    coordinate,
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:40.2f} KM".format(distance_coastline),
    )
)
locations_green_red.add_child(distance_marker)
```

```python
# Create a `folium.PolyLine` object using the coastline coordinates and launch site coordinate
locations = [[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]]
lines=folium.PolyLine(locations= locations, weight=1)
locations_green_red.add_child(lines)
```

```python
# Railway 
launch_site_lat = 28.56327
launch_site_lon = -80.57673
railroad_lat = 28.56287
railroad_lon = -80.58699

distance_railraod = calculate_distance(launch_site_lat, launch_site_lon, railroad_lat, railroad_lon)

coordinate_railroad = [railroad_lat,railroad_lon] 
distance_marker_railroad = folium.Marker(
    coordinate_railroad,
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:40.2f} KM".format(distance_railraod),
    )
)
locations_green_red.add_child(distance_marker_railroad)

locations = [[launch_site_lat, launch_site_lon], [railroad_lat, railroad_lon]]
lines=folium.PolyLine(locations= locations, weight=1)
locations_green_red.add_child(lines)
```
